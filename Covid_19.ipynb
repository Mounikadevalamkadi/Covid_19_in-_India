## Step4: Exploratory Data Analysis(EDA)

## 1.Boxplot for confirmrd Cases by State

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='State/UnionTerritory', y='Confirmed', data=df)
plt.title('Distribution of Confirmed Cases by State/Union Territory')
plt.xticks(rotation=90)
plt.show()
```

## 2.Pie Chart for Proportion of Cases (Confirmed, Cured, Deaths)

```python
cases = df[['Confirmed', 'Cured', 'Deaths']].sum()
cases.index = ['Confirmed', 'Cured', 'Deaths']

plt.figure(figsize=(8, 8))
plt.pie(cases, labels=cases.index, autopct='%1.1f%%', startangle=140, colors=['Skyblue', 'lightgreen', 'purple'])
plt.title('Proportion of Total Cases')
plt.show()
```

## 3.Heatmap for State-wise Correlation of Metrics

```python
plt.figure(figsize=(12, 8))
correlation = df.groupby('State/UnionTerritory')[['Confirmed', 'Cured', 'Deaths']].sum().corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('State-wise Correlation of Metrics')
plt.show()
```

## 4.Recovery Rate by State with Subset

```python
df['RecoveryRate'] = df['Cured'] / df['Confirmed']
top_states = df.groupby('State/UnionTerritory')['Confirmed'].sum().nlargest(5).index
subset_df = df[df['State/UnionTerritory'].isin(top_states)]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='State/UnionTerritory', y='RecoveryRate', data=subset_df)
plt.title('Recovery Rate by Top States')
plt.xticks(rotation=90)
plt.ylabel('Recovery Rate')
plt.show()
```

## 5.Time Series Analysis

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming daily data, decompose the time series
result = seasonal_decompose(df.groupby('Date')['Confirmed'].sum(), model='additive', period=30)

# Plot the decomposition
result.plot()
plt.show()
```

## 6.Deaths Vs Confirmed Cases

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Confirmed', y='Deaths', data=df)
plt.title('Confirmed Cases vs. Deaths')
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.show()
```

## 7.State-wise Recovery Rate

```python
plt.figure(figsize=(12,8))
statewise_recovery = df.groupby('State/UnionTerritory')['RecoveryRate'].mean().sort_values(ascending=False)
sns.barplot(x=statewise_recovery.values, y=statewise_recovery.index)
plt.title('State-wise Recovery Rate')
plt.xlabel('Recovery Rate')
plt.ylabel('State/Union Territory')
plt.show()
```

## 8.Heatmap of Cases by Time

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%I:%M %p').dt.hour
hourly_cases = df.groupby(['Hour'])['Confirmed'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(hourly_cases.pivot_table(index='Hour', values='Confirmed'), cmap='YlGnBu', annot=True)
plt.title('Heatmap of Cases by Hour of the Day')
plt.show()
```

## Step5: Feature Engineering

## Active Cases

```python
df['ActiveCases'] = df['Confirmed'] - (df['Cured'] + df['Deaths'])
```

## Daily Deaths

```python
df['DeathsPerDay'] = df.groupby('Date')['Deaths'].diff().fillna(df['Deaths'])
```

## Recent Recovery Improvement

```python
df['RecentRecoveryImprovement'] = df.groupby('State/UnionTerritory')['Cured'].rolling(window=7).mean().reset_index(0, drop=True)
```

```python
# Fill or drop missing values
df['RecentRecoveryImprovement'].fillna(df['RecentRecoveryImprovement'].mean(), inplace=True)
```
